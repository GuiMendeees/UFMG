```python
#Consulta 1
#Liste todos os codigos e nomes de funcoes da subfuncao que possui o maior numero de funcoes.
```

```python
df = pd.read_sql_query("SELECT * FROM FUNCAO WHERE COD_FUNCAO IN "
+"(SELECT COD_FUNCAO FROM DETALHAMENTO_FUNCAO WHERE COD_SUBFUNCAO = "
+"(SELECT COD_SUBFUNCAO FROM DETALHAMENTO_FUNCAO GROUP BY COD_SUBFUNCAO ORDER BY COUNT(*) DESC ))"
+"ORDER BY COD_FUNCAO",conn)
```

```python
#Consulta 5
#Liste o codigo e nome dos orgaos que possuiram os 5 maiores valores de despesas totais
#nas funcoes de educacao e saude durante o ano de 2018.
```

```python
df = pd.read_sql_query("SELECT COD_ORGAO, NOME_ORGAO FROM ORGAO NATURAL JOIN"
+"(SELECT COD_ORGAO, SUM(VALOR_DESPESA) FROM DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO WHERE COD_FUNCAO IN "
+"(SELECT COD_FUNCAO FROM FUNCAO WHERE NOME_FUNCAO IN ('Educação', 'Saúde'))" 
+" AND ANO_DESPESA = 2018 GROUP BY COD_ORGAO ORDER BY SUM(VALOR_DESPESA) DESC)",conn)
df
```

```python
#Consulta 7
#Liste os programas de governo e a quantidade de orgaos governamentais distintos envolvidos
#nas despesas publicas desses programas. Nesta consulta, desconsidere valores
#de despesa negativos ou iguais a zero no perIodo.
```

```python
df = pd.read_sql_query("SELECT DISTINCT NOME_PROGRAMA_GOVERNO, COUNT(COD_ORGAO) AS QNT_ORGAO FROM "
+"DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE VALOR_DESPESA > 0 GROUP BY COD_ORGAO "
+"ORDER BY QNT_ORGAO DESC",conn)
df
```

```python
#Consulta 10
#Liste todas as despesas publicas (codigo da despesa, mes e ano da despesa, modalidade,valor), 
#ordenadas de forma decrescente pelo valor, em que o programa de governo eH aquele que possui 
#o maior valor total em despesas durante o ano de 2018.
```

```python
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, COD_DESPESA, MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, "
+"VALOR_DESPESA FROM DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE ANO_DESPESA = 2018 ORDER BY VALOR_DESPESA DESC",conn)
df
```

```python
#Consulta 4
#Liste o mes e ano, modalidade, programa de governo e valor de todas as despesas do
#orgao Ministerio da Educacao no primeiro semestre de 2018.
```

```python
df = pd.read_sql_query("SELECT MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA,NOME_PROGRAMA_GOVERNO, "
+"VALOR_DESPESA FROM DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE COD_ORGAO = "
+"(SELECT COD_ORGAO FROM ORGAO WHERE NOME_ORGAO = 'Ministério da Educação') AND MES_DESPESA <7",conn)
df
```

```python
#Consulta 11 
#Liste os codigos e nomes dos orgaos governamentais relacionados ao programa de
#governo (ou programas, caso haja empate) que alcancou mais registros de despesas
#publicas durante o ano.
```

```python
df = pd.read_sql_query("SELECT COD_ORGAO, NOME_ORGAO,COUNT(VALOR_DESPESA) AS NUM_DESPESAS, SUM(VALOR_DESPESA) "
+"AS TOTAL FROM (SELECT VALOR_DESPESA, COD_ORGAO, NOME_ORGAO FROM DESPESA_PUBLICA NATURAL JOIN ORGAO)GROUP BY "
+"COD_ORGAO ORDER BY NUM_DESPESAS DESC, TOTAL DESC",conn)
df

```
